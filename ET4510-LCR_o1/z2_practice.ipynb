```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([0, 2, 4, 6, 8])
plt.plot(a, marker='o')
plt.title("Simple Array Plot")
plt.xlabel("Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
a = np.array([[1, 2], [3, 4]])
plt.imshow(a, cmap='viridis')
plt.colorbar()
plt.title("Heatmap of 2D Array")
plt.show()
```

```python
# This is a module file named car_module.py (module: a file grouping code)
class Car:  # Class: blueprint for creating car objects
    def __init__(self, make, model):  # Method: function inside the class
        self.make = make      # Attribute: data belonging to the object
        self.model = model    # Attribute
    
    def drive(self):  # Method, describes behavior of the object
        print(f"The {self.make} {self.model} is driving.")

# Create an instance (object) of the Car class
my_car = Car("Tesla", "Model S")
my_car.drive()  # Call method on the instance

print(my_car.make)  # Access attribute of the instance
print(f"'Hi, Jason!'")
print(f'"Hi, Jason!"')
```
